# filtering the data frame in condn

```python
import pandas as pd
```

```python
df_laptops = pd.read_csv( 'https://raw.githubusercontent.com/Subhas333/Project-1/main/laptop_price.csv', encoding='latin-1' )
```

```python
df_laptops.head(3)
```

# dataframe in one condn

```python
import numpy as np
```

```python
np.where(df_laptops['Price_euros']>2000,'expensive','cheap')
#here we done check price and update the true with expensive and false for cheap
```

```python
#set it in new column
df_laptops['price_category'] = np.where(df_laptops['Price_euros']>2000,'expensive','cheap')
```

```python
df_laptops.head(3)
```

```python
df_laptops.value_counts('price_category')
```

# filtering data on multiple condition

```python
# checking apple company laptop more than 2000 euros
# using and
df_laptops[(df_laptops['Company']=='Apple') & (df_laptops['Price_euros']>2000)]
```

```python
# using or operator |
df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Price_euros']>2000)]
```

```python
#  new one condn
df_new=df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')& (df_laptops['Price_euros']>2000)]
```
