```python
import pandas as pd
import numpy as np
```

```python
df_laptops = pd.read_csv('https://raw.githubusercontent.com/Subhas333/Project-1/main/laptop_price.csv', encoding='latin-1')
```

```python
df_laptops.head()
```

###create condn and value variables on +2 choices


```python
condn=[df_laptops['Price_euros']>3000,
       (df_laptops['Price_euros']>2000 ) &(df_laptops['Price_euros']<3000),
       (df_laptops['Price_euros']>1000 ) &(df_laptops['Price_euros']<2000),
                                           df_laptops['Price_euros']<1000

        ]
values= ['too expensive','ecpensive','Affortable','cheap']
```

```python
df_laptops['Price_category'] = np.select(condn, values, default='Unknown')
```

```python
df_laptops
```

```python
df_laptops['Price_category'].value_counts()
```

# exercise : inch size =too big ,big,small,too small

```python
condn= [
    df_laptops['Inches']>=16,
    (df_laptops['Inches']>=14) & (df_laptops['Inches']<16),
    (df_laptops['Inches']>=12) & (df_laptops['Inches']<14),
    df_laptops['Inches']<12
]
values= ['too big','big','small','too small']
```

```python
df_laptops['Inch_category'] = np.select(condn, values, default='Unknown')
```

```python
df_laptops.head()
```

```python
df_laptops['Inch_category'].value_counts()
```
