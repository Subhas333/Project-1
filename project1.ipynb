# multiple links
1.   https://www.football-data.co.uk/mnz4281/2122/E0.csv
2.   https://www.football-data.co.uk/mnz4281/2122/E2.csv
3.   https://www.football-data.co.uk/mnz4281/2122/E3.csv
4.   https://www.football-data.co.uk/mnz4281/2122/E1.csv

```python
# create list of leagues
leagues = ['E0', 'E2', 'E3', 'E1' ]
frame=[]
for league in leagues:
    df = pd.read_csv(f'https://www.football-data.co.uk/mnz4281/2122/{league}.csv')
    frame.append(df)
```

```python
len(frame)
```

```python
frame[0]
```

```python
root = "https://www.football-data.co.uk/mnz4281"
```

##looping through leagues and read multiple csv and append it into list

```python
for league in leagues:
  for season in range(15,21):
    df = pd.read_csv(f'{root}/{season}{season+1}/{league}.csv')
    frame.append(df)
```
